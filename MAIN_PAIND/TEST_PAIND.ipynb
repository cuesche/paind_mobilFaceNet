```python
import tensorflow as tf

# Laden der Protobuf-Datei
with tf.io.gfile.GFile('C:/Users/Marku/OneDrive/Desktop/Studium/Semester_5/Paind/Testsumgebung/paind_mobilFaceNet/arch/pretrained_model/MobileFaceNet_9925_9680.pb', 'rb') as f:
    graph_def = tf.compat.v1.GraphDef()
    graph_def.ParseFromString(f.read())

# Importieren des Graphens
with tf.compat.v1.Session() as sess:
    # Restaurieren des Graphens
    sess.graph.as_default()
    tf.import_graph_def(graph_def, name='')
    # Jetzt können Sie den Graphen für Ihre Zwecke verwenden
```

```python
for op in sess.graph.get_operations():
    print(op.name)
```

```python
# Eingabe-Tensor
input_tensor = sess.graph.get_tensor_by_name('img_inputs:0')
# Ausgabe-Tensor
output_tensor = sess.graph.get_tensor_by_name('embeddings:0')

print(input_tensor)
print(output_tensor)
```

```python


# Instanzieren Sie den Konverter mit dem Pfad zur .pb-Datei und den Input-/Output-Tensornamen
converter = tf.lite.TFLiteConverter.from_frozen_graph(
    graph_def_file='C:/Users/Marku/OneDrive/Desktop/Studium/Semester_5/Paind/Testsumgebung/paind_mobilFaceNet/arch/pretrained_model/MobileFaceNet_9925_9680.pb',  # Pfad zur .pb-Datei
    input_arrays=[input_tensor],  # Name des Input-Tensors
    output_arrays=[output_tensor]  # Name des Output-Tensors
)

# Setzen Sie die Konvertierungsoptionen, falls notwendig
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# Konvertieren Sie das Modell
tflite_model = converter.convert()

# Schreiben Sie das konvertierte Modell in eine Datei
with open('model.tflite', 'wb') as f:
    f.write(tflite_model)
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

# Lade den Frozen Graph
with tf.gfile.GFile('C:/Users/Marku/OneDrive/Desktop/Studium/Semester_5/Paind/Testsumgebung/paind_mobilFaceNet/arch/pretrained_model/MobileFaceNet_9925_9680.pb', 'rb') as f:
    graph_def = tf.GraphDef()
    graph_def.ParseFromString(f.read())

# Importiere den GraphDef in einen neuen Graphen
with tf.Graph().as_default() as graph:
    tf.import_graph_def(graph_def, name='')

# Konvertiere in ein SavedModel
with tf.Session(graph=graph) as sess:
    tf.saved_model.simple_save(
        session=sess,
        export_dir='C:/Users/Marku/OneDrive/Desktop/Studium/Semester_5/Paind/Testsumgebung/paind_mobilFaceNet/arch/pretrained_model/a',
        inputs={'input_tensor': graph.get_tensor_by_name('img_inputs:0')},
        outputs={'output_tensor': graph.get_tensor_by_name('embeddings:0')}
    )

```
